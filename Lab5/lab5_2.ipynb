```python
from Memristor import Memristor
```

```python
SM2_TO_M2 = ((10**(-2))**2)
NM_TO_M = 10**(-9)
```

```python
D = 10 * NM_TO_M
mu = (10**(-10)) * SM2_TO_M2
nu = 100
w0 = D / 10
R_ON = 1
R_OFF = R_ON * 160
U0 = 1
```

```python
def U_sin(t, U0, nu):
    return U0 * np.sin(nu * t)

def Phi_sin(t, U0, nu):
    return U0 * (1 - np.cos(nu * t)) / nu
```

```python
def U_sin_sqr(t, U0, nu):
    return U0 * (np.sin(nu * t) ** 2)

def Phi_sin_sqr(t, U0, nu):
    return U0 * (2 * nu * t - np.sin(2 * nu * t)) / (4 * nu)
```

```python
m_sin_sqr1 = Memristor(D, mu, nu, w0, 1, 20, lambda t: U_sin_sqr(t, U0, nu), lambda t: Phi_sin_sqr(t, U0, nu))
m_sin_sqr2 = Memristor(D, mu, nu, w0, 1, 380, lambda t: U_sin_sqr(t, U0, nu), lambda t: Phi_sin_sqr(t, U0, nu))
```

```python
m_sin_sqr1.task1(plt, 1000, 5)
m_sin_sqr1.task2(plt, 1000, 5)
m_sin_sqr1.task3(plt, 1000, 5)
```
