```python
from Memristor import Memristor
```

```python
SM2_TO_M2 = ((10**(-2))**2)
NM_TO_M = 10**(-9)
```

```python
D = 10 * NM_TO_M
mu = (10**(-10)) * SM2_TO_M2
nu = 100
w0 = D / 10
R_ON = 1
R_OFF = R_ON * 160
U0 = 1
```

```python
def U_sin(t, U0, nu):
    return U0 * np.sin(nu * t)

def Phi_sin(t, U0, nu):
    return U0 * (1 - np.cos(nu * t)) / nu
```

```python
nu1 = 1
m_sin1 = Memristor(D, mu, nu1, w0, 1, 160, lambda t: U_sin(t, U0, nu1), lambda t: Phi_sin(t, U0, nu1))
nu2 = 10
m_sin2 = Memristor(D, mu, nu2, w0, 1, 160, lambda t: U_sin(t, U0, nu2), lambda t: Phi_sin(t, U0, nu2))
nu3 = 100
m_sin3 = Memristor(D, mu, nu3, w0, 1, 160, lambda t: U_sin(t, U0, nu3), lambda t: Phi_sin(t, U0, nu3))
nu4 = 1000
m_sin4 = Memristor(D, mu, nu4, w0, 1, 160, lambda t: U_sin(t, U0, nu4), lambda t: Phi_sin(t, U0, nu4))
```
