```python
from Memristor import Memristor
```

```python
SM2_TO_M2 = ((10**(-2))**2)
NM_TO_M = 10**(-9)
```

```python
D = 10 * NM_TO_M
mu = (10**(-10)) * SM2_TO_M2
nu = 100
w0 = D / 10
R_ON = 1
R_OFF = R_ON * 160
U0 = 1
```

```python
def U_sin(t, U0, nu):
    return U0 * np.sin(nu * t)

def Phi_sin(t, U0, nu):
    return U0 * (1 - np.cos(nu * t)) / nu
```

```python
m_sin1 = Memristor(D, mu, nu, w0, 160, 1, lambda t: U_sin(t, 1, nu), lambda t: Phi_sin(t, 1, nu))
m_sin2 = Memristor(D, mu, nu, w0, 160, 1, lambda t: U_sin(t, 2, nu), lambda t: Phi_sin(t, 2, nu))
m_sin3 = Memristor(D, mu, nu, w0, 160, 1, lambda t: U_sin(t, 4, nu), lambda t: Phi_sin(t, 4, nu))
```

```python
m_sin1.task1(plt, 1000, 5)
```
